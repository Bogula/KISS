```python
# Import der module for text 
# to speech conversion und Ausgabe

import IPython
from gtts import gTTS
  
import os
```

```python



  
# Text, den man ausgeben will
#mytext = 'Welcome to Artificial Intelligence Center Hamburg!'
  
mytext = "Hallo wie geht es euch heute abend? Ich hoffe, der Kurs hat euch ein wenig Spa√ü gemacht."

# Sprache
language = 'de'
  
# Slow oder fast

myobj = gTTS(text=mytext, lang=language, slow=False)
  

# in Datei schreiben mp3
myobj.save("sagwas.mp3")

#Datei ausgeben
IPython.display.Audio('sagwas.mp3', autoplay=True)
```
